```python
import numpy as np
from seasonal_stats import PlayerSeasonals

with open ('top10k.txt', 'r') as f:

    players = f.read().splitlines()

    f.close()
playerScores = []
for i in range(0,10000):
    playerScores.append(PlayerSeasonals(players[i]))
    print i
```

```python
wins = [player.wins for player in playerScores]
hours = [player.total for player in playerScores]
print hours
```

```python
array_wins = np.array(wins)
array_hours = np.array(hours)
array_playerScores = np.array(playerScores)
mask = np.not_equal(wins, np.zeros(len(wins)))
print mask
array_wins = array_wins[mask]
array_hours = array_hours[mask]
array_playerScores = array_playerScores[mask]
```

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams.update({'figure.autolayout': True})
```

```python
mask2 = np.greater_equal(array_wins, 3*np.ones(59))
print mask2
threeWins = array_playerScores[mask2]
threeWinshrs = array_hours[mask2]
threeWinswins = array_wins[mask2]
```

```python

fig = plt.figure(figsize=(24,12))
ax = fig.add_subplot(111)
labels = [player.name for player in threeWins]
plt.semilogx(hours,wins, 'o', c='b', alpha=0.5)
plt.semilogx(array_hours,array_wins,'o')
plt.axis([0,2000,0,20])
for label, x, y in zip(labels, threeWinshrs, threeWinswins):
    if np.random.rand() <= 0.5:
        sign = -1
    else:
        sign = 1
    plt.annotate(
    label, 
    xy = (x, y), xytext = (sign*30, sign*30*np.random.rand()+np.random.randint(5,10)),
    textcoords = 'offset points', ha = 'right', va = 'bottom',
    bbox = dict(boxstyle = 'round,pad=0.5', fc = 'yellow', alpha = 0.5),
    arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc3,rad=0')
    ,size=8)
plt.title('Seasonal Hiscore Results of the top {} ranked players'.format(len(playerScores)), size=16)
plt.xlabel('Hours spent (log scale)')
plt.ylabel('Number of Hiscores Won')
plt.savefig('Seasonal Hiscores')
plt.show()
```
